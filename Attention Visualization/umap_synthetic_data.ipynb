```python
import umap
import umap.plot

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure

import random
from umap import UMAP
```

## Generate synthetic data

```python
# Create synthetic data
np.random.seed(19680801)

def rgb_to_hex(rgb):
    return '%02x%02x%02x' % rgb

fig = plt.figure(figsize = [8, 8], dpi = 150)

ax = fig.add_subplot(projection='3d')

n = 1000

r = list(np.random.normal(100, 25, n))
g = list(np.random.normal(100, 25, n))
b = list(np.random.normal(100, 25, n))

data_3d = []
for i in range(n):
    data_3d.append([r[i], g[i], b[i]])

color_points = []
for i in range(500):
    color_points.append('#b5332a')
for i in range(500):
    color_points.append('#1f55de')
ax.scatter(r, g, b, c = color_points)
```

```python
umap_euc = UMAP(n_components = 3, init = 'random', random_state = 0, metric = "euclidean")
embedding_2d_euc = umap_euc.fit_transform(data_3d)

# run 2d embedding for different hyperparameters and display results

figure(figsize=(6, 6), dpi=150)

x = list(embedding_2d_euc[:, 0])
y = list(embedding_2d_euc[:, 1])

#create scatterplot
scatter = plt.scatter(x, y, c=color_points, s = 2)

plt.tick_params(bottom = False, left = False)
plt.xticks([]) 
plt.yticks([]) 
plt.show()

umap_cosine = UMAP(n_components = 3, init = 'random', random_state = 0, metric = "cosine")
embedding_2d_cosine = umap_cosine.fit_transform(data_3d)

# run 2d embedding for different hyperparameters and display results

figure(figsize=(6, 6), dpi=150)

x = list(embedding_2d_cosine[:, 0])
y = list(embedding_2d_cosine[:, 1])

#create scatterplot
scatter = plt.scatter(x, y, c=color_points, s = 2)

plt.tick_params(bottom = False, left = False)
plt.xticks([]) 
plt.yticks([]) 
plt.show()
```

## Scaling

```python
# SCALING

sums1 = [0, 0, 0]
for i in range(500):
    r[i] /= 2
    g[i] /= 2
    b[i] /= 2
    sums1[0] += r[i]
    sums1[1] += g[i]
    sums1[2] += b[i]

sums2 = [0, 0, 0] 
for i in range(500, 1000):
    r[i] *= 2
    g[i] *= 2
    b[i] *= 2
    sums2[0] += r[i]
    sums2[1] += g[i]
    sums2[2] += b[i]
    
data_3d = []
for i in range(n):
    data_3d.append([r[i], g[i], b[i]])

fig = plt.figure(figsize = [10, 10], dpi = 300)
ax = fig.add_subplot(projection='3d')
ax.scatter(r, g, b, c = color_points)
```

```python
umap_euc = UMAP(n_components = 3, init = 'random', random_state = 0, metric = "euclidean")
embedding_2d_euc = umap_euc.fit_transform(data_3d)

# run 2d embedding for different hyperparameters and display results

figure(figsize=(6, 6), dpi=150)

x = list(embedding_2d_euc[:, 0])
y = list(embedding_2d_euc[:, 1])

#create scatterplot
scatter = plt.scatter(x, y, c=color_points, s = 2)

plt.tick_params(bottom = False, left = False)
plt.xticks([]) 
plt.yticks([]) 
plt.show()

umap_cosine = UMAP(n_components = 3, init = 'random', random_state = 0, metric = "cosine")
embedding_2d_cosine = umap_cosine.fit_transform(data_3d)

# run 2d embedding for different hyperparameters and display results

figure(figsize=(6, 6), dpi=150)

x = list(embedding_2d_cosine[:, 0])
y = list(embedding_2d_cosine[:, 1])

#create scatterplot
scatter = plt.scatter(x, y, c=color_points, s = 2)

plt.tick_params(bottom = False, left = False)
plt.xticks([]) 
plt.yticks([]) 
plt.show()
```

## Scaling and Translation

```python
# compute centroid differences
diffs = []
for i in range(3):
    diffs.append((sums2[i] - sums1[i])/500)
    
# Perform translation
for i in range(500):
    r[i] += diffs[0]
    g[i] += diffs[1]
    b[i] += diffs[2]

data_3d = []
for i in range(n):
    data_3d.append([r[i], g[i], b[i]])

fig = plt.figure(figsize = [10, 10], dpi = 300)
ax = fig.add_subplot(projection='3d')
ax.scatter(r, g, b, c = color_points)
```

```python
umap_euc = UMAP(n_components = 3, init = 'random', random_state = 0, metric = "euclidean")
embedding_2d_euc = umap_euc.fit_transform(data_3d)

# run 2d embedding for different hyperparameters and display results

figure(figsize=(6, 6), dpi=150)

x = list(embedding_2d_euc[:, 0])
y = list(embedding_2d_euc[:, 1])

#create scatterplot
scatter = plt.scatter(x, y, c=color_points, s = 2)

plt.tick_params(bottom = False, left = False)
plt.xticks([]) 
plt.yticks([]) 
plt.show()

umap_cosine = UMAP(n_components = 3, init = 'random', random_state = 0, metric = "cosine")
embedding_2d_cosine = umap_cosine.fit_transform(data_3d)

# run 2d embedding for different hyperparameters and display results

figure(figsize=(6, 6), dpi=150)

x = list(embedding_2d_cosine[:, 0])
y = list(embedding_2d_cosine[:, 1])

#create scatterplot
scatter = plt.scatter(x, y, c=color_points, s = 2)

plt.tick_params(bottom = False, left = False)
plt.xticks([]) 
plt.yticks([]) 
plt.show()
```
